```python
import numpy as np
import pandas as pd

scores = np.random.randint(50, 100, size=(50,5))
scores_df = pd.DataFrame(scores,
                         columns=['國文','英文','數學','地理','歷史'],
                         index=range(1,51))
scores_df
sum_value = scores_df.sum(axis=1)
mean_value = scores_df.mean(axis=1)
scores_df['總分'] = sum_value
scores_df['平均'] = mean_value
```

```python
scores_df
```

```python
#排序總分 sort
scores_df.sort_values(by='總分') 
#axis預設值為0，按照欄位排序(上下垂直)，可以不用寫。
#ascending預設值為True，總分由小至大。
```

```python
#總分由大排至小。
#找出分數最低的五名
scores_df.sort_values(by='總分', ascending=False).iloc[-5:]
```

```python
scores_df['平均'] < 70 #比較運算子，傳出布林值。False: >=70；True" <70。
```

```python
#使用subscript[]，過濾資料
scores_df[scores_df['平均'] < 80]
```

```python
#有幾位同學平均分數小於80
scores_df[scores_df['平均'] < 80].count()
```

```python
#使用subscribe[]的搜尋方式
#平均分數小於80，同時歷史不及格
scores_df[(scores_df['平均'] < 80) & (scores_df['歷史'] < 60)]
```

```python
#使用query()的方式搜尋
scores_df.query('平均<80')
```

```python
scores_df.query('平均<80' and '歷史<60')
```
