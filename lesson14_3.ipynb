```python
import numpy as np
import pandas as pd

scores = np.random.randint(50, 100, size=(50,5))
scores_df = pd.DataFrame(scores,
                         columns=['國文','英文','數學','地理','歷史'],
                         index=range(1,51))
scores_df
sum_value = scores_df.sum(axis=1)
mean_value = scores_df.mean(axis=1)
scores_df['總分'] = sum_value
scores_df['平均'] = mean_value
```

```python
scores_df
```

```python
scores_df
scores_df.style.highlight_between(left=50,right=60)
```

```python
scores_df
scores_df.style\
    .format(precision=2)\
    .highlight_between(left=50,right=59)\
    .to_excel('abc.xlsx')

#\:表示這一行跟下一行接在一起
#.:用句點代表逗號
```
